## Aggregate Sales and Profit by sub-Category with Profit Margin

```python
grouped_sc_df = df.groupby('Sub-Category').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
#grouped_df = df.groupby(['Sub-Category', 'Category']).agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
grouped_sc_df['ProfitMargin'] = (grouped_sc_df['Profit']/grouped_sc_df['Sales']) * 100
print("Aggregate data with Profit Margin for sub-category")
print(grouped_sc_df)
```

## Aggregate Sales and Profit by Region with Gross Profit Margin

```python
grouped_region_df = df.groupby('Region').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
#grouped_df = df.groupby(['Sub-Category', 'Category']).agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
grouped_region_df['ProfitMargin'] = (grouped_region_df['Profit']/grouped_region_df['Sales']) * 100
print("Aggregate data with Profit Margin for Region")
print(grouped_region_df)
```

The data shows that the Western region generated the most sales and profits, followed by the Eastern region. While the Central region had more sales than the Southern region, the Southern region outperformed the Central region in terms of profit

```python
# Calculating the total sales and profit by region with Profit Margin
region_total_sales = grouped_region_df['Sales'].sum()
region_total_profit = grouped_region_df['Profit'].sum()
region_overall_profit_margin = (region_total_profit / region_total_sales) * 100
```

```python
print("\nTotal Sales and Profit Across All Region")
print(f"Total Sales: ${region_total_sales:,.2f}")
print(f"Total Profit: ${region_total_profit:,.2f}")
print(f"Overall Profit Margin: {region_overall_profit_margin:.2f}%")
```

## Aggregate Sales and Profit by Category with Gross Profit Margin

```python
grouped_cat_df = df.groupby('Category').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
#grouped_df = df.groupby(['Sub-Category', 'Category']).agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
grouped_cat_df['ProfitMargin'] = (grouped_cat_df['Profit']/grouped_cat_df['Sales']) * 100
print("Aggregate data with Gross Profit Margin for category")
print(grouped_cat_df)
```

The data shows that Technology had the highest sales and generated the most profit. Although Furniture had more sales than Office Supplies, but Office Supplies achieved a higher profit than Furniture.

```python
# Calculating the total sales and profit by category with Profit Margin
cat_total_sales = grouped_cat_df['Sales'].sum()
cat_total_profit = grouped_cat_df['Profit'].sum()
cat_overall_profit_margin = (cat_total_profit / cat_total_sales) * 100
```

```python
print("\nTotal Sales and Profit Across All Categories")
print(f"Total Sales: ${cat_total_sales:,.2f}")
print(f"Total Profit: ${cat_total_profit:,.2f}")
print(f"Overall Profit Margin: {cat_overall_profit_margin:.2f}%")
```

The data shows the Total Sales, Total Profit, and Profit Margin for the Superstore. Note: With an overall Profit Margin of 12.47%, the store demonstrates good profitability, though there are areas that can be improved to enhance performance further.

## Calculating Total Sales and Profit by Ship Mode for a Specific State

```python
def get_sales_profit_by_ship_mode(df, state_of_interest):
    
    filtered_df = df[df['State'] == state_of_interest]
    sales_profit_by_ship_mode = filtered_df.groupby('Ship Mode').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
    sales_profit_by_ship_mode.columns = ['Ship Mode', 'TotalSales', 'TotalProfit']
    return sales_profit_by_ship_mode
```

```python
get_sales_profit_by_ship_mode(df, 'New York')
```

- In terms of profit, New York outperformed California in First Class and Same Day shipping modes, while California had higher profits in Second Class and Standard Class modes, but the profit difference in Standard Class was small relative to the sales difference.
